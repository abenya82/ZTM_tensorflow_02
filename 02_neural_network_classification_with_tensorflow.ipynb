```python
# visualize the model's predictions:
#  this function will 
#   take in a trained model, features, and labels
#  Create a mesh grid (numpy) of the different X values
# make predictions across the meshgrid
# plot the predictions and a line between classes
```

```python
import numpy as np

def plot_decision_boundary(model,X,y):
    # Plots the decision boundary created by a model predicting on X
    x_min, x_max = X[:,0].min() - 0.1, X[:,0].max() + 0.1
    y_min, y_max = X[:,1].min() - 0.1, X[:,1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    # create X values
    x_in = np.c_[xx.ravel(), yy.ravel()]

    # make predictions
    y_pred = model.predict(x_in)

    # Check for multi-class
    if len(y_pred[0])>1:
        print("doing multiclass classification")
        #we have to reshape our prediction to get them ready for plotting
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("doing binary classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    #plot the decision boundary
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:,0], X[:,1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    
```

## Finding the best learning rate

### Learning Rate callback
- extra piece of functionality, you can add to your model while it is running
- need a modified loss curves plot


```python
#set random seed
tf.random.set_seed(42)

#build a model
model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation='relu'),
    tf.keras.layers.Dense(4,activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

#compile the model
model_9.compile(loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy'])

# Create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))
## every epoch, traverse a set of learning rate values, starting at 1-e4, 
## increasing by 10**(epoch/20), every epoch


#Fit the model
history_9 = model_9.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])
```

```python
# check the history
pd.DataFrame(history_9.history).plot(figsize=(10,7), xlabel='epochs')
```

```python
# different looking loss curves
```

```python
# plot the learning rate vs. the loss
#here are the learning rates that get used:
lrs = 1e-4 * (10 ** (tf.range(100)/20))       #here, 100 is number of epochs
len(lrs)
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_9.history["loss"])
plt.xlabel("learning rate")
plt.ylabel("Loss")
```

```python
#take learning rate where the curve is still initially decreasing, but before it starts to level out
```

## more evaluation metrics

# Accuracy
- tp+tn / tp+tn+fp+fn
- tf.keras.metrics.Accuracy()
- sklearn.metrics.accuracy_score()
- default for classification problems, not best for imbalanced classes

# Precision
- tp/tp+fp
- tf.keras.metrics.Precision()
- sklearn.metrics.precision_score()
- Higher precision leads to fewer false positives

# Recall
- tp/ tp+fn
- tf.keras.metrics.Recall()
- sklearn.metrics.recall_score()
- Higher Recall leads to fewer false negatives

# F1-score
- 2 \* (precision \* recall)/precision+recall
- sklearn.metrics.f1_score()
- combination of precision and recall, good overall metrics for classification

# Confusion Matrix
- Custom function in tf
- sklearn.metrics.confusion_matrix()
- when comparing predictions to truth labels to see where model gets confused.
- Can be hard to use with large numbers of classes

### Precision/Recall trade off

## Evaluating a multiclass classification model

- Evaluate performance with confusion matrix
- Assess some of its predictions
- Improve its results
- Save and export

```python
#create confusion matrix
import itertools
from sklearn.metrics import confusion_matrix

def make_confusion_matrix(y_true,y_pred,classes=None, figsize=(10,10), text_size=15):


    
    # Create the confusion matrix
    cm = confusion_matrix(y_true,tf.round(y_pred))
    cm_norm = cm.astype("float") / cm.sum(axis=1)[:,np.newaxis]  #normalize confusion matrix
    n_classes = cm.shape[0]  #binary classificaion, so we expect size = 2

    # Let's prettify it

    fig, ax = plt.subplots(figsize=figsize)
    #Create Matrix Plot
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    fig.colorbar(cax)
    # set labels to be classes

    if classes:
        labels = classes
    else:
        labels = np.arange(cm.shape[0])
    # Let's label the axis
    ax.set(title='confusion matrix',
            xlabel="predicted label",
            ylabel='True Label',
            xticks=np.arange(n_classes),
            yticks=np.arange(n_classes),
            xticklabels=labels,
            yticklabels=labels)
    # set x axis labels to bottom
    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()

    #adjust label size
    ax.yaxis.label.set_size(text_size)
    ax.xaxis.label.set_size(text_size)
    ax.title.set_size(text_size)


    #set threshold for different colors
    threshold = (cm.max() + cm.min())/2.

    #plot the text on each cell
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i, f"{cm[i,j]} ({cm_norm[i,j]*100:.1f}%)",
            horizontalalignment="center",
            color="white" if cm[i,j] > threshold else "black",
            size=text_size)
```

# Lets make a function that
- Plots random image
- Makes prediction on that image
- Label the plot with the truth label & the predicted label

```python
import random
def plot_random_image(model, images, true_labels, classes):

    #Set random int
    i = random.randint(0,len(images))

    #create predictions and targets
    target_image = images[i]
    pred_probs = model.predict(target_image.reshape(1,28,28))
    pred_label = classes[pred_probs.argmax()]
    true_label = classes[true_labels[i]]

    #Plot the image
    plt.imshow(target_image, cmap=plt.cm.binary)

    # change the color of the titles depending on if the prediction is right
    if pred_label == true_label:
        color="green"
    else:
        color="red"
    
    #Add xlabel information
    plt.xlabel("Pred:{} {:2.0f}% (True: {})".format(pred_label,100*tf.reduce_max(pred_probs), true_label),color=color)
```

```python
plot_random_image(model=model_14, 
                    images=test_data_norm,
                    true_labels=test_labels,
                    classes=class_names)
```
